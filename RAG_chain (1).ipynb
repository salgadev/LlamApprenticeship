```python
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader
```

```python


# Define document directory (replace with your data path)
document_dir = "path/"

# Load documents
reader = SimpleDirectoryReader(input_files=[document_dir])
documents = reader.load_data()

# Create vector store index
index = VectorStoreIndex.from_documents(documents, show_progress=True)

# Function to query the index
def query_documents(query):
  """
  This function takes a query string and retrieves relevant documents.
  """
  query_engine = index.as_query_engine()
  results = query_engine.query(query)

  # Access top retrieved document
  top_result = results[0]
  document_id = top_result["document_id"]
  document_content = documents[document_id]["text"]

  # Print retrieved document information
  print(f"Top retrieved document for query '{query}':")
  print(f"Document ID: {document_id}")
  print(f"Content snippet:")
  print(document_content[:100] + "...")  # Print first 100 characters

# Example usage
query = "What are the benefits of using cloud computing?"
query_documents(query)
```
